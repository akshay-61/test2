```python
def crop(path1,path2):
    #share the path to images 
    path1
    #share the path to json files 
    path2
    os.chdir(Home)
    #os.makedirs(os.path.join("static", "test_crop_final"))
    Path("test_crop_final").mkdir(exist_ok=True)
    #read images infolder
    for filename1 in os.listdir(path1):
        for filename2 in os.listdir(path2):
            if filename1.endswith('.png'):
                img_file = path1 + "\\" + filename1
                img = cv2.imread(img_file)
                if filename2.endswith('.json'):
                    if filename1[:-4]==filename2[:-5] :
                        json_file = path2 + "\\" + filename2
                        df = pd.read_json(json_file)
                        # getting number of coloumns
                        length = len(list(df.columns))
                        # getting number of rows
                        count_row = df.shape[0]
                        List= []
                        for i in range(count_row):
                            for j in range(length):
                                try:
                                    test= df.iloc[i][j]
                                    List.append(test)
                                except IndexError:
                                    pass
                        try:
                            img1 = img[List[2]:List[2]+List[6], List[0]:List[0]+List[4]]
                            img2 = img[List[3]:List[3]+List[7], List[1]:List[1]+List[5]]
                            cv2.imwrite("test_crop_final" + "/" + os.path.basename(img_file[:-4]) +"_1"+".png", img1)
                            cv2.imwrite("test_crop_final" + "/" + os.path.basename(img_file[:-4]) +"_2"+".png", img2)
                        except IndexError:
                                    pass
crop(path1=UPLOAD_FOLDER,path2=Text_detection_json)
```

```python
def OCR(path):            
    # text extraction
    os.chdir(Home)
    path
    Path("easyocr_crop_3").mkdir(exist_ok=True)
    for filename in os.listdir(path):
        if filename.endswith('.png'):
            img_file = path + "\\" + filename
            reader = easyocr.Reader(['en'])
            result = reader.readtext(img_file)
            List= []
            length= len(result)
            for i in range(length):
                try:
                    List.append(result[i][1])
                except IndexError:
                    pass 
            # dictionary of lists  
            dict = {'Final_easyocr': List}  
            df = pd.DataFrame(dict)
            one_string = ' '.join(df['Final_easyocr'].tolist())
            data = io.StringIO(one_string)
            df_2 = pd.read_csv(data, sep=",")
            # saving the dataframe 
            df_2.to_csv("easyocr_crop_3" + "/" + os.path.basename(img_file[:-4]) +".csv")
OCR(path=r'C:\Users\shak1001\Desktop\streamlit\pasted_final')
```
